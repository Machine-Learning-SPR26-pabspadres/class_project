```python
import pandas as pd
```

```python
df = pd.read_excel("/Users/pablopadres/Documents/School/MS2/thesis/Data/raw_data/human_rights_risk_&_responsiveness.xlsx")

df.head()
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")

df.columns
```

```python
HRD_POINTS = 35
OTHER_POINTS = 20
REQUEST_POINTS = 20
RESPONSE_POINTS = 25
```

```python
max_hrd = df["HRD_Attacks"].max()
max_other = df["Other_Allegations"].max()
max_requests = df["Response_Requests"].max()
max_response = 100

max_hrd, max_other, max_requests
```

```python
df["hrd_deduction"] = (df["HRD_Attacks"] / max_hrd) * HRD_POINTS
df["other_allegations_deduction"] = (df["Other_Allegations"] / max_other) * OTHER_POINTS
df["response_requests_deduction"] = (df["Response_Requests"] / max_requests) * REQUEST_POINTS
```

```python
df["response_rate_deduction"] = ((max_response - df["Response_Rate"]) / max_response) * RESPONSE_POINTS
```

```python
df["HRRR_score"] = (
    100
    - df["hrd_deduction"]
    - df["other_allegations_deduction"]
    - df["response_requests_deduction"]
    - df["response_rate_deduction"]
)
```

```python
df["HRRR_score"] = df["HRRR_score"].clip(lower=0).round(2)

df[[
    "Brand",
    "HRD_Attacks",
    "Other_Allegations",
    "Response_Requests",
    "Response_Rate",
    "HRRR_score"
]]
```
